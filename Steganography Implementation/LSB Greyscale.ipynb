```python
import cv2
import pystegano
import os
from IPython.display import Image, display
import requests
```

```python
# Documentation
help(pystegano.lsb_grayscale)
```

```python
# Generate random Lorem Ipsum text
def random_message():
    response = requests.get("http://loripsum.net/api/10/headers/plaintext")
    if response.status_code == 200:
        secret_msg = response.text
    else:
        # Default secret message, when requests fail
        secret_msg = "Hello There!"
    return secret_msg
```

```python
# Set directory of cover images
cover_dirname = "C:/Users/PC/Masters Files/SteganographyDetection/Datasets/Dataset Cover Images"
# Set working directory
dirname = "C:/Users/PC/Masters Files/SteganographyDetection/Datasets/Dataset LSB Greyscale"
```

```python
# Add stego payload
for i, filename in enumerate(os.listdir(cover_dirname)):
    cover_image = cv2.imread(cover_dirname + "/" + filename, cv2.IMREAD_UNCHANGED)
    secret_message = random_message()
    encoded_image = pystegano.lsb_grayscale.encode(cover_image, secret_message)
    cv2.imwrite(dirname + "/" + filename, encoded_image)
```

```python
# Display the stego image
display(Image(dirname + "/59.png"))
```

```python
# Load stego image
stego_image = cv2.imread(dirname + "/59.png", cv2.IMREAD_UNCHANGED)
```

```python
# Attempt to decode the secret message
decoded_message = pystegano.lsb_grayscale.decode(stego_image)
```
