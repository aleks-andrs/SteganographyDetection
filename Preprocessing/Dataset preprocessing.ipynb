```python
import os
import shutil
import cv2
from PIL import Image
import matplotlib.pyplot as plt
%matplotlib inline
```

```python
# Original image location
original_dir = "C:/Users/PC/Masters Files/SteganographyDetection/Datasets/Dataset Original"
# Cover image location
cover_dir = "C:/Users/PC/Masters Files/SteganographyDetection/Datasets/Dataset Cover Images"
```

```python
# Copy all JPG images from Dataset Original to Dataset Cover Images
for i, filename in enumerate(os.listdir(original_dir)):
    shutil.copy(original_dir + "/" + filename, cover_dir)
```

```python
# Resize images
for i, filename in enumerate(os.listdir(cover_dir)):
    image = Image.open(cover_dir + "/" + filename)
    image.thumbnail((400, 400))
    image.save(cover_dir + "/" + filename)
    image.close()
```

```python
# Rename JPG images "0" to "59" and save as PNG
for i, filename in enumerate(os.listdir(cover_dir)):
    os.rename(cover_dir + "/" + filename, cover_dir + "/" + str(i) + ".png")
```

```python
# Display original image for testing purposes
cover_image = cv2.cvtColor(cv2.imread(cover_dir + "/59.png"),cv2.COLOR_BGR2RGB)
plt.imshow(cover_image)
plt.show()
```

```python
# Crop images
for i, filename in enumerate(os.listdir(cover_dir)):
    cover_image = cv2.imread(cover_dir + "/" + filename, cv2.IMREAD_UNCHANGED)
    y=50
    x=50
    h=80
    w=80
    crop_img = cover_image[y:y+h, x:x+w]
    cv2.imwrite(cover_dir + "/" + filename, crop_img)
```

```python
# Display image after cropping
cover_image = cv2.cvtColor(cv2.imread(cover_dir + "/59.png"),cv2.COLOR_BGR2RGB)
plt.imshow(cover_image)
plt.show()
```
